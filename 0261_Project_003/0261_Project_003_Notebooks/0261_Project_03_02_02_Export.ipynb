```python
### TTSplit

from sklearn.model_selection import train_test_split
X_train, X_valid_to_be, y_train, y_valid_to_be = train_test_split(X, y,
                                                                  test_size=0.2,
                                                                  train_size=None,
                                                                  random_state=42,
                                                                  shuffle=True,
                                                                  stratify=None)

X_valid, X_test, y_valid, y_test = train_test_split(X_valid_to_be, y_valid_to_be,
                                                    test_size=0.5,
                                                    train_size=None,
                                                    random_state=42,
                                                    shuffle=True,
                                                    stratify=None)

```

```python
## Random Forest Elbow Method

from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score,classification_report, confusion_matrix, ConfusionMatrixDisplay



error_rate_model_forest_elbow_01_valid = []
error_rate_model_forest_elbow_01_test = []

for n in range(1, 101):
    model_forest_elbow_01 = RandomForestClassifier(n_estimators=n,
                                                   criterion='gini',
                                                   max_depth=None,
                                                   min_samples_split=2,
                                                   min_samples_leaf=1,
                                                   min_weight_fraction_leaf=0.0,
                                                   max_features='sqrt',
                                                   max_leaf_nodes=None,
                                                   min_impurity_decrease=0.0,
                                                   bootstrap=True,
                                                   oob_score=False,
                                                   n_jobs=None,
                                                   random_state=None,
                                                   verbose=0,
                                                   warm_start=False,
                                                   class_weight=None,
                                                   ccp_alpha=0.0,
                                                   max_samples=None)
    
    model_forest_elbow_01.fit(X_train, y_train)

    model_forest_elbow_01_valid_preds = model_forest_elbow_01.predict(X_valid)
    acc_score_model_forest_elbow_01_valid = accuracy_score(y_true=y_valid, y_pred=model_forest_elbow_01_valid_preds, normalize=True, sample_weight=None)
    error_model_forest_elbow_01_valid = 1 - acc_score_model_forest_elbow_01_valid
    error_rate_model_forest_elbow_01_valid.append(error_model_forest_elbow_01_valid)

    model_forest_elbow_01_test_preds = model_forest_elbow_01.predict(X_test)
    acc_score_model_forest_elbow_01_test = accuracy_score(y_true=y_test, y_pred=model_forest_elbow_01_test_preds, normalize=True, sample_weight=None)
    error_model_forest_elbow_01_test = 1 - acc_score_model_forest_elbow_01_test
    error_rate_model_forest_elbow_01_test.append(error_model_forest_elbow_01_test)

    #from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay

```

```python
## Random Forest Elbow Method - ranges from 1 to 5 estimators and see Classification report

from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, f1_score, classification_report, confusion_matrix, ConfusionMatrixDisplay

accuracy_model_forest_elbow_03_valid = []
accuracy_model_forest_elbow_03_test = []

error_rate_model_forest_elbow_03_valid = []
error_rate_model_forest_elbow_03_test = []

f1_score_model_forest_elbow_03_valid = []
f1_score_model_forest_elbow_03_test = []

for n in range(1, 6):
    model_forest_elbow_03 = RandomForestClassifier(n_estimators=n,
                                                   criterion='gini',
                                                   max_depth=None,
                                                   min_samples_split=2,
                                                   min_samples_leaf=1,
                                                   min_weight_fraction_leaf=0.0,
                                                   max_features='sqrt',
                                                   max_leaf_nodes=None,
                                                   min_impurity_decrease=0.0,
                                                   bootstrap=True,
                                                   oob_score=False,
                                                   n_jobs=None,
                                                   random_state=None,
                                                   verbose=0,
                                                   warm_start=False,
                                                   class_weight=None,
                                                   ccp_alpha=0.0,
                                                   max_samples=None)
    
    model_forest_elbow_03.fit(X_train, y_train)

    model_forest_elbow_03_valid_preds = model_forest_elbow_03.predict(X_valid)
    acc_score_model_forest_elbow_03_valid = accuracy_score(y_true=y_valid, y_pred=model_forest_elbow_03_valid_preds, normalize=True, sample_weight=None)

    accuracy_model_forest_elbow_03_valid.append(acc_score_model_forest_elbow_03_valid)

    error_model_forest_elbow_03_valid = 1 - acc_score_model_forest_elbow_03_valid
    error_rate_model_forest_elbow_03_valid.append(error_model_forest_elbow_03_valid)
    
    f1_score_valid = f1_score(y_true=y_valid, y_pred=model_forest_elbow_03_valid_preds, labels=None, pos_label=1, average='binary', sample_weight=None, zero_division='warn')
    f1_score_model_forest_elbow_03_valid.append(f1_score_valid)
    
    
    model_forest_elbow_03_test_preds = model_forest_elbow_03.predict(X_test)
    acc_score_model_forest_elbow_03_test = accuracy_score(y_true=y_test, y_pred=model_forest_elbow_03_test_preds, normalize=True, sample_weight=None)
    
    accuracy_model_forest_elbow_03_test.append(acc_score_model_forest_elbow_03_test)
    
    error_model_forest_elbow_03_test = 1 - acc_score_model_forest_elbow_03_test
    error_rate_model_forest_elbow_03_test.append(error_model_forest_elbow_03_test)
    
    f1_score_test = f1_score(y_true=y_test, y_pred=model_forest_elbow_03_test_preds, labels=None, pos_label=1, average='binary', sample_weight=None, zero_division='warn')
    f1_score_model_forest_elbow_03_test.append(f1_score_valid)

df_model_forest_elbow_03_results = pd.DataFrame(data=[accuracy_model_forest_elbow_03_valid,
                                                   error_rate_model_forest_elbow_03_valid,
                                                   f1_score_model_forest_elbow_03_valid,
                                                   accuracy_model_forest_elbow_03_test,
                                                   error_rate_model_forest_elbow_03_test,
                                                   f1_score_model_forest_elbow_03_test],
                                             index=['acc_score_model_forest_elbow_03_valid',
                                                    'error_rate_model_forest_elbow_03_valid',
                                                    'f1_score_model_forest_elbow_03_valid',
                                                    'acc_score_model_forest_elbow_03_test',
                                                    'error_rate_model_forest_elbow_03_test',
                                                    'f1_score_model_forest_elbow_03_test'])



df_model_forest_elbow_03_results = df_model_forest_elbow_03_results.transpose()

df_model_forest_elbow_03_results


plt.figure(figsize=(6,4))
plt.scatter(df_model_forest_elbow_03_results['acc_score_model_forest_elbow_03_valid'].index, df_model_forest_elbow_03_results['acc_score_model_forest_elbow_03_valid'], label=['Validation'])
plt.scatter(df_model_forest_elbow_03_results['acc_score_model_forest_elbow_03_test'].index, df_model_forest_elbow_03_results['acc_score_model_forest_elbow_03_test'], label=['Test'])

plt.legend(loc='best')
plt.title("acc_score_model_forest_elbow_03")


plt.figure(figsize=(6,4))
plt.scatter(df_model_forest_elbow_03_results['error_rate_model_forest_elbow_03_valid'].index, df_model_forest_elbow_03_results['error_rate_model_forest_elbow_03_valid'], label=['Validation'])
plt.scatter(df_model_forest_elbow_03_results['error_rate_model_forest_elbow_03_test'].index, df_model_forest_elbow_03_results['error_rate_model_forest_elbow_03_test'], label=['Test'])

plt.legend(loc='best')
plt.title("error_rate_model_forest_elbow_03")


plt.figure(figsize=(6,4))
plt.scatter(df_model_forest_elbow_03_results['f1_score_model_forest_elbow_03_valid'].index, df_model_forest_elbow_03_results['f1_score_model_forest_elbow_03_valid'], label=['Validation'])
plt.scatter(df_model_forest_elbow_03_results['f1_score_model_forest_elbow_03_test'].index, df_model_forest_elbow_03_results['f1_score_model_forest_elbow_03_test'], label=['Test'])

plt.legend(loc='best')
plt.title("f1_score_model_forest_elbow_03")



plt.show()
```
